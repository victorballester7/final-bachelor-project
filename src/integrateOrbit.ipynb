```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
# Define constants
GM = 398600.4415e+9

# Define function to calculate acceleration due to gravity


def a_gravity(r, t):
    x, y, z, vx, vy, vz = r
    r = np.array([x, y, z])
    v = np.array([vx, vy, vz])
    a = -GM * r / (np.linalg.norm(r)**3)
    return np.concatenate((v, a))


# Define initial conditions
r0 = [3314301.34844, 5927810.41895, 9194.18738817, -4152.48524775,
      2315.02428879, 6007.55530593]  # [x, y, z, vx, vy, vz]

# Define time array
T=5000
t = np.linspace(0, T, 1000)
```

```python
# Integrate orbit using odeint
r = odeint(a_gravity, r0, t)
```

```python
# Plot orbit trajectory

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(r[:, 0], r[:, 1], r[:, 2], 'r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Orbit Trajectory')
plt.show()
```
