```python
# convert Astronomical Almanac data to x, y, z in TEME frame
from skyfield.api import load
import numpy as np
```

```python
ts = load.timescale()
filenames = [
    [0, "../../data/almanac_position/moon"],
    [1, "../../data/almanac_position/sun"],
    [2, "../../data/almanac_position/venus"],
    [3, "../../data/almanac_position/mars"],
    [4, "../../data/almanac_position/jupiter"],
]
end_inp = "_almanac.txt"
end_out = ".txt"
```

```python
for filename in filenames:
  with open(filename[1] + end_inp, "r") as f:
    lines = f.readlines()
  L = []
  for epoch in lines:
    year = int(epoch[0:4])
    month=1
    if epoch[5:8]=="Jan":
        month = 1
    elif epoch[5:8]=="Feb":
        month=2
    day = int(epoch[9:11])
    hour = int(epoch[12:14])
    minute = int(epoch[15:17])
    second = float(epoch[18:22])
    ra_hours = int(epoch[27:29])
    ra_minutes = int(epoch[30:32])
    ra_seconds = float(epoch[33:39])
    d = int(epoch[46:48])
    if(d>9):
      dec_degrees = int(epoch[44]+epoch[46:48])
    else:
      dec_degrees = int(epoch[44]+epoch[47:48])
    dec_minutes = int(epoch[49:51])
    dec_seconds = float(epoch[52:57])
    distance = float(epoch[63:])

    dec = dec_degrees + dec_minutes / 60 + dec_seconds / 3600
    ra = ra_hours * 15 + ra_minutes * 0.25 + ra_seconds * 0.004166666666666667

    mjd_utc = ts.utc(year, month, day, hour, minute, second).tt - 2400000.5

    
    x = distance * np.cos(np.deg2rad(dec)) * np.cos(np.deg2rad(ra))
    y = distance * np.cos(np.deg2rad(dec)) * np.sin(np.deg2rad(ra))
    z = distance * np.sin(np.deg2rad(dec))

    if (filename[0] == 0):
      x *= 1000
      y *= 1000
      z *= 1000
    else: # we convert to km
      x *= 149597870700
      y *= 149597870700
      z *= 149597870700

    L.append([mjd_utc, x, y, z])
  
  with open(filename[1] + end_out, "w") as f:
    for i in L:
      f.write(str(i[0]) + " " + str(i[1]) + " " + str(i[2]) + " " + str(i[3]) + "\n")
```
