```python
# in order to compute the variation of the TLE angles over time.
import numpy as np
```

```python
filename = "../data/teme/ISS_tles.txt"
with open(filename, "r") as f:
  lines = f.readlines()
```

```python
L=[]
for i in range(0, len(lines)):
    L.append([float(lines[i].split()[j]) for j in range(7)])
```

```python
# compute the orbital elements:
r = np.array([L[i][1:4] for i in range(len(L))])
v = np.array([L[i][4:7] for i in range(len(L))])
h = np.cross(r, v)
n = np.cross(np.array([0, 0, 1]), h)
GM = 398600.4415e+9
e_vec = np.cross(v, h)/GM - r/np.linalg.norm(r, axis=1)[:, None]
e = np.linalg.norm(e_vec, axis=1)
i = np.arccos(h[:, 2]/np.linalg.norm(h, axis=1))
Omega = np.arccos(n[:, 0]/np.linalg.norm(n, axis=1))
if (n[:, 1] < 0).any():
    Omega[n[:, 1] < 0] = 2*np.pi - Omega[n[:, 1] < 0]
# omega = np.arccos(np.sum(n*e_vec, axis=1)/(np.linalg.norm(n, axis=1)*e))
omega = np.arccos(np.dot(n, e_vec.T)/(np.linalg.norm(n, axis=1)*e))
if (e_vec[:, 2] < 0).any():
    omega[e_vec[:, 2] < 0] = 2*np.pi - omega[e_vec[:, 2] < 0]
# nu = np.arccos(np.dot(e_vec, r.T)/(e*np.linalg.norm(r, axis=1)))
# if (np.dot(r, v.T) < 0).any():
#     nu[np.dot(r, v.T) < 0] = 2*np.pi - nu[np.dot(r, v.T) < 0]
a = np.linalg.norm(h, axis=1)**2/(GM*(1 - e**2))

# Convert to degrees
i = np.rad2deg(i)
Omega = np.rad2deg(Omega)
omega = np.rad2deg(omega)
```
